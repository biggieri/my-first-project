```python
# Lists
products = ["Apple", "Milk", "Bread", "Cheese", "Eggs"]
prices = [1.99, 3.49, 2.99, 5.99, 4.49]

print(products)
print(prices)

# Access a single item - just like referencing a cell in Excel
print(products[0])  # First item
print(products[2])  # Third item
```

```python
for product in products:
    print(f"Product: {product}")
```

```python
for i in range(len(products)):
    if prices[i] > 3.00:
        print(f"{products[i]} is expensive: ${prices[i]}")
    else:
        print(f"{products[i]} is affordable: ${prices[i]}")
```

```python
# Define a function
def calculate_total(price, quantity):
    total = price * quantity
    return total

# Use the function
print(calculate_total(3.49, 80))
print(calculate_total(5.99, 45))
print(calculate_total(1.99, 150))
```

```python
quantities = [150, 80, 60, 45, 90]

def calculate_total(price, quantity):
    return round(price * quantity, 2)
for i in range(len(products)):
    total = calculate_total(prices[i], quantities[i])
    print(f"{products[i]}: ${total}")
```

```python
# A dictionary
product = {
    "name": "Milk",
    "category": "Dairy",
    "price": 3.49,
    "quantity": 80
}

print(product["name"])
print(product["price"])
print(product["quantity"])
```

```python
# List of dictionaries - like a table
products_data = [
    {"name": "Apple",  "category": "Produce", "price": 1.99, "quantity": 150},
    {"name": "Milk",   "category": "Dairy",   "price": 3.49, "quantity": 80},
    {"name": "Bread",  "category": "Bakery",  "price": 2.99, "quantity": 60},
    {"name": "Cheese", "category": "Dairy",   "price": 5.99, "quantity": 45},
    {"name": "Eggs",   "category": "Dairy",   "price": 4.49, "quantity": 90}
]

# Loop through each row
for product in products_data:
    total = calculate_total(product["price"], product["quantity"])
    print(f"{product['name']} ({product['category']}): ${total}")
```
