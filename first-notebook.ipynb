```python
import numpy as np

# Create a messy dataset with missing values
messy_data = [
    {"name": "Apple",  "category": "Produce", "price": 1.99,  "quantity": 150},
    {"name": "Milk",   "category": "Dairy",   "price": None,  "quantity": 80},
    {"name": "Bread",  "category": None,       "price": 2.99,  "quantity": 60},
    {"name": "Cheese", "category": "Dairy",   "price": 5.99,  "quantity": None},
    {"name": "Eggs",   "category": "Dairy",   "price": 4.49,  "quantity": 90}
]

messy_df = pd.DataFrame(messy_data)
messy_df
```

```python
messy_df.isnull()
```

```python
messy_df.isnull().sum()
```

```python
# Option 1 — Drop rows with any missing values
cleaned_df = messy_df.dropna()
cleaned_df
```

```python
# Option 2 — Fill missing values with something meaningful
filled_df = messy_df.copy()
filled_df["price"] = filled_df["price"].fillna(0)
filled_df["quantity"] = filled_df["quantity"].fillna(0)
filled_df["category"] = filled_df["category"].fillna("Unknown")
filled_df
```

```python
# First table - products and their category
products_df = pd.DataFrame([
    {"product_id": 1, "name": "Apple",  "category": "Produce"},
    {"product_id": 2, "name": "Milk",   "category": "Dairy"},
    {"product_id": 3, "name": "Bread",  "category": "Bakery"},
    {"product_id": 4, "name": "Cheese", "category": "Dairy"},
    {"product_id": 5, "name": "Eggs",   "category": "Dairy"}
])

# Second table - pricing information
pricing_df = pd.DataFrame([
    {"product_id": 1, "price": 1.99, "quantity": 150},
    {"product_id": 2, "price": 3.49, "quantity": 80},
    {"product_id": 3, "price": 2.99, "quantity": 60},
    {"product_id": 4, "price": 5.99, "quantity": 45},
    {"product_id": 6, "price": 2.49, "quantity": 100}
])

print("Products table:")
print(products_df)
print("\nPricing table:")
print(pricing_df)
```

```python
inner = pd.merge(products_df, pricing_df, on="product_id", how="inner")
print("Inner Join:")
print(inner)
```

```python
# Left join - keep all products, fill missing pricing with NaN
left = pd.merge(products_df, pricing_df, on="product_id", how="left")
print("Left join:")
print(left)
```

```python
# Left join - keep all products, fill missing pricing with NaN
left = pd.merge(products_df, pricing_df, on="product_id", how="left")
print("Left join:")
print(left)
```

```python
# Outer join - keep everything from both tables
outer = pd.merge(products_df, pricing_df, on="product_id", how="outer")
print("Outer join:")
print(outer)
```
